```python
import os
import pandas as pd
import sys
import subprocess
import re
import json
```

```python
def remove_include_macros_namesapce(source_code: str):
    return re.sub(r'#\s*include\s*[<"].*[>"].*\n', '', source_code)

def c_preprocess(source_code: str):
    source_code = remove_include_macros_namesapce(source_code)

    with open('tmp.cpp', 'w') as f:
        f.write(source_code)

    terminal = r'''gcc -E tmp.cpp | awk '/# [0-9]* "tmp.cpp"/ {skip=0; next} /# [0-9]* ".*"/ {skip=1} (skip<1) {print}' '''
    i_text = subprocess.check_output(terminal, shell=True, timeout=0.1).decode(sys.stdout.encoding)
    return i_text
```

```python
DATA_PATH = './data/submissions/'
FILE_NAMEs = os.listdir(DATA_PATH)
```

```python
er_contest = []
for fn in FILE_NAMEs[:]:
    df = pd.read_csv(open(DATA_PATH + fn, 'r'), encoding='utf-8', engine='c')

    df = df[['problem', 'submission_id', 'source_code']]
    try:
        df['source_code'] = df['source_code'].apply(lambda x: c_preprocess(x))
        df.to_csv('./data/cpreprocess/' + fn, index=False)
    except:
        er_contest.append(fn)
        continue
```
